```python
with open('election.json','r') as fp:
    Ele = json.load(fp)
    
```

```python
Ele
```

```python
Data = []
PTY = []
for item in Ele:
    thread = ""
    for key,value in item.items():
        if key == "Pradesh":
            thread = value
        if key == "District":
            thread = thread+"-"+value
        if key == "District_sub":
            thread = thread+"-"+value
        if key == "Party":
            thread = thread+"-"+value
            if value[0:3] != "Ind":
                PTY.append(value)
        if key == "value":
            v = value
    print(thread,v)
    Data.append({"id" : thread,"value":v})
```

```python
DF = pd.DataFrame(Data)
DF = DF.set_index("id")
DF.to_csv("ndata.csv")
```
